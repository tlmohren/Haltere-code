```python
import numpy as np
from sympy import *
import scipy.integrate as integrate
from scipy.integrate import quad
import matplotlib.pyplot as plt
from sympy.utilities.lambdify import lambdify
import scipy
import pandas as pd

t = Symbol('t')
phi = pi/2*sin(t)
dphi = diff(phi,t)*2/pi
ddphi = diff(dphi,t)
```

```python


df = pd.DataFrame 
funcs = [cos(phi), cos(phi)**2, dphi**2, cos(phi)*sin(phi), dphi*sin(phi) ]
accel = ['Euler e_theta', 'centrifugal dphi e_r','centrifugal dtheta e_r ','centrifugal e_phi','Coriolis e_theta']

n_elem = 11
dfA = pd.DataFrame(index = range(n_elem),columns =funcs)
dfB = pd.DataFrame(index = range(n_elem),columns =funcs)
for funTemp in funcs:
    for j in range(n_elem):
        funC= lambdify(t,funTemp*cos(j*t))
        funS= lambdify(t,funTemp*sin(j*t))
        Atemp = quad( funC, -np.pi, np.pi)
        Btemp = quad( funS, -np.pi, np.pi)
        dfA.loc[j,funTemp] = np.round( Atemp[0]/(np.pi), 5)
        dfB.loc[j,funTemp] = np.round( Btemp[0]/(np.pi), 5)
```

```python
for figs in range(5):
    n_func = figs
    lam_t = lambdify(t,funcs[n_func], modules=['numpy'])

    t_vals = np.linspace(0,2* np.pi, 100)
    y_vals = lam_t(t_vals)

    f = dfA[funcs[n_func]][0]/2
    for j in np.arange(1,n_elem):
        f = f+ dfA[funcs[n_func]][j]*cos(j*t) + dfB[funcs[n_func]][j]*sin(j*t)
    f_lam = lambdify(t,f, modules=['numpy'])
    yf_vals = f_lam(t_vals)

    plt.plot(t_vals, y_vals, label= accel[n_func])
    plt.ylabel("Y")
    plt.legend(loc='right')
plt.show()
```
